```python
# split dfbreastcancer dataset to x and y so we can setup in the model
xdfbreastcancer=dfbreastcancer[["radius_mean","texture_mean","smoothness_mean","compactness_mean","concavity_mean"]]
ydfbreastcancer=dfbreastcancer["diagnosis"]
```

```python
# import preprocessing to use normalize function to normlize x to fit it in the model
from sklearn import preprocessing
xdfbreastcancer_norm = preprocessing.normalize(xdfbreastcancer)
```

```python
dfbreastcancer= dfbreastcancer.astype('category')
```

```python
#change row of 'diagnosis' to 0 and 1 instead of M and B
dfbreastcancer['diagnosis']=dfbreastcancer['diagnosis'].cat.codes
```

```python
#to show 'diagnosis' after convert it to 0 and 1 instead of M and B
dfbreastcancer['diagnosis']
```

```python
#import train_test_split to divide dfbreastcancer dataset to train and test 
from sklearn.model_selection import train_test_split
```

```python
#assign x tain y train as 60% and x test y test as 40%
x_traind,x_testd,y_traind,y_testd=train_test_split(xdfbreastcancer,ydfbreastcancer,test_size=0.4,random_state=42, stratify=ydfbreastcancer)
```

```python
#import KNN-Classifier
from sklearn.neighbors import KNeighborsClassifier

#Setup arrays from 1 to 9 to store training and test accuracies
neighbors = npmining.arange(1,9)
traind_accuracy =npmining.empty(len(neighbors))
testd_accuracy = npmining.empty(len(neighbors))

for m,k in enumerate(neighbors):
    #Setup a knn classifier with k neighbors
    knn = KNeighborsClassifier(n_neighbors=k)
    
    #Fit xandy train to the model
    knn.fit(x_traind, y_traind)
    
    #Compute accuracy for training set
    traind_accuracy[m] = knn.score(x_traind, y_traind)
    
    #Compute accuracy for test set
    testd_accuracy[m] = knn.score(x_testd, y_testd) 
```

```python
#to show best neighbors for the model 
plt.title('k-NN number of neighbors from the loop')
plt.plot(neighbors, testd_accuracy, label='Test Accuracy')
plt.plot(neighbors, traind_accuracy, label='Train accuracy')
plt.legend()
plt.xlabel('neighbors')
plt.ylabel('the Accuracy')
plt.show()
```

```python
#knn classifier with 7 neighbors depnding in best neighbors figure
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
#Fit xy train in the model
knn.fit(x_traind,y_traind)
```

```python
#show accurcy for xy test.
knn.score(x_testd,y_testd)
```

```python
#show accurcy for xy train to copmare it with xy test.
knn.score(x_traind, y_traind)
```

```python
#let us get the predictions using the classifier we had fit above
y_predd = knn.predict(x_testd)
```

```python
#import confusionn_matrix
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_testd,y_predd)
sns.heatmap(mat.T, square=True, annot=True, fmt='d' , cbar=False)
plt.title('confusion matrix')
plt.xlabel('True classes')
plt.ylabel('predction class')
```
